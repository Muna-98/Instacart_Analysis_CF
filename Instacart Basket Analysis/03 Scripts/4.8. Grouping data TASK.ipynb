```python
# 2. Find the aggregated mean of the “order_number” column grouped by “department_id”
merged_ords_prods.groupby('department_id').agg({'order_number':['mean']})
```

#### Here we can observe all 21 departments with the mean value of orders. We can also notice that the difference between the lowest and highest order numbers is larger in this case.

```python
# 4. Creating a loyalty flag for existing customers using the transform() and loc() functions.
merged_ords_prods['max_order']=merged_ords_prods.groupby(['customer_id'])['order_number'].transform(np.max)
merged_ords_prods.head()
```

```python
merged_ords_prods.loc[merged_ords_prods['max_order']>40,'loyalty_flag']='Loyal customer'
merged_ords_prods.loc[(merged_ords_prods['max_order']<=40)&(merged_ords_prods['max_order']>10), 'loyalty_flag']='Regular customer'
merged_ords_prods.loc[merged_ords_prods['max_order']<=10, 'loyalty_flag']='New customer'
```

```python
merged_ords_prods['loyalty_flag'].value_counts(dropna=False)
```

```python
# 5.check the basic statistics of the product prices for each loyalty category
merged_ords_prods.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

```python
# 6. Create a spending flag for each user based on the average price across all their orders 
#Finding average spending per customer
merged_ords_prods['average_price']=merged_ords_prods.groupby(['customer_id'])['prices'].transform(np.mean)
merged_ords_prods.head()
```

```python
merged_ords_prods.shape
```

```python
#Creating flags
merged_ords_prods.loc[merged_ords_prods['average_price']<10, 'spending_flag']='Low spender'
merged_ords_prods.loc[merged_ords_prods['average_price']>=10, 'spending_flag']='High spender'
merged_ords_prods['spending_flag'].value_counts(dropna=False)
```

```python
pd.options.display.max_rows = None
merged_ords_prods[['customer_id', 'average_price', 'spending_flag']].head(100)
```

```python
#Creating an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.
merged_ords_prods['median_order_days']=merged_ords_prods.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
merged_ords_prods[['customer_id','median_order_days']].head(50)
```

```python
merged_ords_prods.loc[merged_ords_prods['median_order_days']>20, 'order_frequency_flag']='Non-frequent customer'
merged_ords_prods.loc[(merged_ords_prods['median_order_days']>10)&(merged_ords_prods['median_order_days']<=20), 'order_frequency_flag']='Regular customer'
merged_ords_prods.loc[merged_ords_prods['median_order_days']<=10, 'order_frequency_flag']='Frequent customer'
merged_ords_prods['order_frequency_flag'].value_counts(dropna=False)
```
